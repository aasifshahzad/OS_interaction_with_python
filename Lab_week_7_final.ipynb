```python
#!/usr/bin/env python3

import re
import os
import csv
import operator


#Creating user_statistics.csv

log_count_error = dict()
log_count_info = dict()

with open("syslog.log","r") as file:
  logs = file.readlines()
    
for log in logs:
  search_text = re.search(r"(ERROR) ([\w ]*) \((\w+)\)", log)
  if search_text != None:
    log_count_error[search_text.group(3)]= log_count_error.get(search_text.group(3),0) + 1

for log in logs:
  search_text = re.search(r"((INFO) ([\w ]*) \[(#\d{4,})\] \((\w+)\))", log)
  if search_text != None:
    log_count_info[search_text.group(5)]= log_count_error.get(search_text.group(5),0) + 1

with open('user_statistics.csv', 'w', newline='') as csvfile:
  csvfile_writer = csv.writer(csvfile, delimiter=',',quotechar='|', quoting=csv.QUOTE_MINIMAL)
  csvfile_writer.writerow(["Username", "INFO", "ERROR"])

i = set(log_count_info.keys())
e= set(log_count_error.keys())
z = e-i
d = dict.fromkeys(z,0)
log_count_info.update(d)

log_count_info = sorted(log_count_info.items(),key=operator.itemgetter(0))
log_count_error = sorted(log_count_error.items(),key=operator.itemgetter(0))

username = [i[0] for i in log_count_info]
value1 = [i[1] for i in log_count_info]
value2 = [i[1] for i in log_count_error]
error_info = zip(value1,value2)
final = list(zip(username,error_info))[:8]

for i,j in final:
  with open('user_statistics.csv', 'a', newline='') as csvfile:
    csvfile_writer = csv.writer(csvfile, delimiter=',',quotechar='|', quoting=csv.QUOTE_MINIMAL)
    csvfile_writer.writerow([i,j[0],j[1]])
        
# Counting errors in total for error_message.csv
error_count = dict()
for log in logs:
  search_text = re.search(r"((INFO|ERROR) ([\w ]*))", log)
  error_count[search_text.group(3)]= error_count.get(search_text.group(3),0) + 1
list_error_count = sorted(error_count.items(),key=operator.itemgetter(1), reverse = True)

with open('error_message.csv', 'w', newline='') as csvfile:
  spamwriter = csv.writer(csvfile, delimiter=  ',',quotechar='|', quoting=csv.QUOTE_MINIMAL)
  spamwriter.writerow(["Error","Count"])

for pair in list_error_count:
  key,value = pair
  with open('error_message.csv', 'a', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',quotechar='|', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow([key,value])
```
